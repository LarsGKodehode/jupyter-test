# Testing MongoDB with pymongo

## Setup
1. Get all the required configuration strings

```python
import os

username = os.getenv("MONGO_USERNAME")
print("User:", username)
password = os.getenv("MONGO_PASSWORD")
print("Password:", password)
```

2. Setup and connect the MongoDB Client, the ServerAPI argument ensures all calls through the client uses the stable API version

```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi

client = MongoClient(
    "mongodb+srv://{}:{}@jupytertesting.cezqd59.mongodb.net/?retryWrites=true&w=majority".format(username, password),
    server_api=ServerApi('1'))
```

## Interacting with the database

1. Create or access a database, MongoDB should create non exsistant databases

```python
db = client.test_database
```

2. Create or access a Collection

```python
collection = db.test_collection
```

3. Create a document that we want to store in the database

```python
post = {
    "author": "John Doe",
    "title": "My Diary",
    "content": "Day #1 of the Jupyter Era was rather uneventful. This however is a good thing.",
    "tags": ["Diary", "Jupyter"]
}
```

6. Insert the document into our **post** collection

```python
posts = db.posts
post_id = posts.insert_one(post).inserted_id
```

the trailing *.inserted_id* returns the id generated by MongoDB for our element

4. We can list out our collections with this command

```python
print(db.list_collection_names())
```

5. We can also get a single element from a collection with the *.find_one()* method

```python
from pprint import pprint

pprint(posts.find_one())
```

We are using pprint (pretty print) here for formating the output

6. We can query a collection for a secific id by passing in a dictionary with **{"_id": wanted_id}**

```python
pprint(posts.find_one({"_id": post_id}))
```
